## Importando as Bibliotecas Necess√°rias



```python
import pandas as pd
import numpy as np
```

# **Extract**

## Ler a tabela dados_pessoas.csv

```python
dados = pd.read_csv('dados_pessoas.csv')
```

```python
dados
```

# Ler a tabela media_sal.csv

```python
media_sal = pd.read_csv('media_sal.csv')
```

```python
media_sal
```

# Transform

## Mesclar os dados das tabelas dados_pessoas.csv e media_sal.csv

```python
tabela_final = dados.merge(media_sal,
                           left_on = 'ID',
                           right_on ='ID',
                           how='left' )
```

# Visualizar a Tabela Final

```python
tabela_final
```

# Load

## Convertendo a tabela_final.csv para .xls e salvando

```python
tabela_final.to_excel("tabela_final.xlsx", index=False)
```
